```
import imutils
import cv2
from google.colab.patches import cv2_imshow
```

```
#Loading the input image
image = cv2.imread("avatar.jpg")
(h, w, d) = image.shape
print("width={}, height={}, depth={}".format(w, h, d))
```

```
cv2_imshow(image)
```

```
#Getting the value of image at x=80, y=150
(B, G, R) = image[150, 80]
print("R={}, G={}, B={}".format(R, G, B))
```

```
# extract a 100x100 pixel square ROI (Region of Interest) from the
# input image starting at x=60,y=160 at ending at x=220,y=320
roi = image[60:160, 220:320]
cv2_imshow(roi)
```

```
# resize the image to 200x200px, aspect ratio not considered
resized = cv2.resize(image, (200, 200))
cv2_imshow(resized)
```

```
#Considering aspect ratio with width = 300px and calculating the height
r = 300.0 / w
dim = (300, int(h * r))
resized = cv2.resize(image, dim)
cv2_imshow(resized)
```

```
# Using imutils library instead of manually calculating the aspect ratio
resized = imutils.resize(image, width=300)
cv2_imshow(resized)
```

```
#Rotating the image by 45 degree (clockwise) from center and applying affine warp
center = (w // 2, h // 2)
M = cv2.getRotationMatrix2D(center, -45, 1.0)
rotated = cv2.warpAffine(image, M, (w, h))
cv2_imshow(rotated)
```

```
#Rotation using imutils
rotated = imutils.rotate(image, -45)
cv2_imshow(rotated)
```

```
#Rotation without the image being clipped
rotated = imutils.rotate_bound(image, 45)
cv2_imshow(rotated)
```

```
# Applying a Gaussian blur with a 7x7 kernel to the image to smooth it, used to reduce high frequency noise
blurred = cv2.GaussianBlur(image, (7, 7), 0)
cv2_imshow(blurred)
```

```
# drawing the rectangle box
#x and y values are calculated manually
output = image.copy()
cv2.rectangle(output, (80, 20), (140, 70), (0, 0, 255), 2)
cv2_imshow(output)
```

```
# draw a green 10px (filled in) circle on the image centered at x=155,y=82
output = image.copy()
cv2.circle(output, (155, 82), 10, (0, 255, 0), -1)
cv2_imshow(output)
```

```
# draw a 2px thick red line from x=60,y=20 to x=400,y=200
output = image.copy()
cv2.line(output, (60, 20), (200, 100), (0, 0, 255), 2)
cv2_imshow(output)
```

```
# Putting text on the image
output = image.copy()
cv2.putText(output, "OpenCV Basics 1", (10, 25), 
	cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 200, 175), 2)
cv2_imshow(output)
```
